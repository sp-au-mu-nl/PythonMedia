```python
%%writefile frameindex.py
import numpy as np

def frameindex(framelength, noverlap, signallength):
  sft=framelength-noverlap
  n=np.int16(np.fix((signallength-framelength)/sft+1))
  findex=np.array([np.arange(0,framelength)]).T \
                  +np.array([np.arange(0,n)])*sft
  return(findex)
```

```python
from frameindex import frameindex
y,fs=librosa.load('aiueo8.wav',sr=None)
yframe=y[frameindex(256,64,y.shape[0])]
f,nframe=yframe.shape
nperiod=np.zeros(nframe)
for i in range(nframe):
  ac=np.correlate(yframe[:,i],yframe[:,i],"full")
  pks=signal.argrelmax(ac,order=10)
  pkloc=(pks[0])[(pks[0]>256).nonzero()]
  if pkloc.size==0:
    nperiod[i]=np.nan
  else:
    nperiod[i]=pkloc[np.argmax(ac[pkloc])]-255
plt.plot(nperiod)
```
