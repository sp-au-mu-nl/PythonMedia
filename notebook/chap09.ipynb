```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import IPython
import librosa
```

```python
fs=8000
t=np.arange(0,1,1/fs)
cs=np.exp(2j*np.pi*440*t)
y=np.real(cs)
plt.plot(y[:100])
IPython.display.Audio(data=y,rate=fs)
```

```python
t=np.arange(0,1,1/fs)
a=880
b=440
k=a/2*t**2+b*t
ch=np.real(np.exp(2j*np.pi*k))
IPython.display.Audio(data=ch,rate=fs)
```

```python
t=np.arange(0,1,1/fs)
a=440
b=5
fv=4
k=a*t-b/(2*np.pi*fv)*np.cos(2*np.pi*fv*t)
plt.plot(np.diff(k))
vib=np.real(np.exp(2j*np.pi*k))
IPython.display.Audio(data=vib,rate=fs)
```

```python
y,fs=librosa.load('a-.wav',sr=None)
t=np.arange(0,y.shape[0]/fs,1/fs)
g=t+t**2/2
g=g[:np.argwhere(g>t[-1])[0,0]]
mody=np.interp(g,t,y)
plt.specgram(mody,Fs=fs,NFFT=512,window=np.hanning(512),noverlap=256)
plt.figure()
plt.specgram(y,Fs=fs,NFFT=512,window=np.hanning(512),noverlap=256)
IPython.display.Audio(data=mody,rate=fs)
```
