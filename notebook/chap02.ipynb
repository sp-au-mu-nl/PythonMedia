```python
cap1=cv2.VideoCapture('342960254.m4v')
h=int(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))
w=int(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
nFrame1=int(cap1.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(cap1.get(cv2.CAP_PROP_FPS))
V1=np.zeros((height,width,3,nFrame1),np.uint8)
for k in range(0,nFrame1):
  ret,f=cap1.read()
  if ret:
     V1[:,:,:,k]=f

cap2=cv2.VideoCapture('343025947.m4v')
nFrame2=int(cap2.get(cv2.CAP_PROP_FRAME_COUNT))
V2=np.zeros((height,width,3,nFrame2),np.uint8)
for k in range(0,nFrame2):
  ret,f=cap2.read()
  if ret:
     V2[:,:,:,k]=f
```

```python
N=300
Vmix=np.zeros((h,w,3,nFrame1+nFrame2-N),np.uint8)
Vmix[:,:,:,:nFrame1-N]=V1[:,:,:,:nFrame1-N]
Vmix[:,:,:,nFrame1:]=V2[:,:,:,N:]
for k in np.arange(N):
  Vmix[:,:,:,nFrame1-N+k]=np.uint8((1-k/N)*V1[:,:,:,nFrame1-N+k]+
   k/N*V2[:,:,:,k])
```

```python
cis.mp4write(Vmix,"vmix.mp4",fps)
```

```python
video_file = "vmix.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video=cv2.VideoWriter(video_file,fourcc,fps,(width,height))
for k in np.arange(nFrame1+nFrame2-N):
  video.write(Vmix[:,:,:,k])
video.release()
```
